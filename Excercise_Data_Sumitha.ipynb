```python
#Graph 1: Total work out time between users
plt.title('Total Workout Time')
plt.xlabel('users')
plt.ylabel('time(hrs)')
plot1= plt.bar(workout_dataE.deviceuuid,workout_dataE.Diff)
```

```python
#Graph 2:Calories Burnt Per User
Calories_burnt = exercise_data[['deviceuuid','calorie']]
plt.title("Calorie burnt per User")
plt.xlabel("Users")
plt.ylabel("Calories")
plt.bar(Calories_burnt.deviceuuid,Calories_burnt.calorie)
plt.rcParams['figure.figsize'] = (5,4)
plt.style.use('grayscale')
```

```python
#Conclusion:the first User F/D7 burns more calories than the other user.
```
