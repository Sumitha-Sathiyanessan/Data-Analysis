```python
#finding the difference in time
floors_data['Start']=pd.to_datetime(floors_data['start_time'].values.astype('<M8[ns]'))
floors_data['End']=pd.to_datetime(floors_data['end_time'].values.astype('<M8[ns]'))
floors_data['Diff']=floors_data['End']-floors_data['Start']
floors_data['Diff']  
floors_data['Diff_minutes'] = floors_data['Diff'].dt.components['minutes']
```

```python
floors_data
```

```python
# Graph1: Relationship between time and floors climbed
plt.xlabel("Time Difference in min")
plt.ylabel("Floors Climbed")
plt.title("No of Floors Climbed")
plot1=plt.bar(floors_data.Diff_minutes, floors_data.floor,color='red' )
```

```python
# Conclusion:The above graph indicates the number of floors a person  climbes in minutes, between 0 to 1 min person climbes
#/ 4 floors, to climb maximum 12 floors a person needs 1 to 2 minutes also some times a person takes more time 0f 5 min to 
#/ higher floors.
```

```python
# Graph2: Comparitive Analysis for months
plt.figure(figsize=(15,5))
plot2=plt.plot_date(floors_data.update_time,floors_data.floor,color='green')
plt.xlabel('months')
plt.ylabel('floors climbed')
 
```
