```python
Sleep_data['Start']=pd.to_datetime(Sleep_data['com.samsung.health.sleep.start_time'].values.astype('M8[m]'))
```

```python
Sleep_data['End']=pd.to_datetime(Sleep_data['com.samsung.health.sleep.end_time'].values.astype('M8[m]'))
```

```python
Sleep_data['Diff']=Sleep_data['End']-Sleep_data['Start']
```
