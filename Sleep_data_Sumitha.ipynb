```python
# calculation for finding the difference in time
sleep_data['Start']=pd.to_datetime(sleep_data['Start'].values.astype('<M8[m]'))
sleep_data['End']=pd.to_datetime(sleep_data['End'].values.astype('<M8[m]'))
sleep_data['Diff']=sleep_data['End']-sleep_data['Start']
```

```python
sleep_data['Diff']
```

```python
#indexing the dataframe
index1=sleep_data.index
columns=sleep_data.columns
values=sleep_data.values
```

```python
#creating a subset of data for graphs and calculation
workout_data=sleep_data[['Start','End','Diff','Activity (steps)','Sleep quality']]
workout_data
```

```python
#indexing the subset dataframe
print(workout_data.dtypes)
index2=workout_data.index
```

```python
# Identify the time slept in hours and minutes
workout_data['time(hours)']=workout_data['Diff'].dt.components['hours']
workout_data['minutes(slept)']=workout_data['Diff'].dt.components['hours']*60+(workout_data['Diff'].dt.components['minutes'])
workout_data
```

```python
# Graph 1: Histogram for hours of sleep/no of days 
plt.xlabel('hours of sleep')
plt.ylabel('no of days')
plot1=plt.hist(workout_data['time(hours)'],bins=20, label='Sleeping Time', color='purple',rwidth=0.95)
```

```python
# Conclusion
A person sleeps 5 hrs per day  in maximum noi  of the days 
```

```python
# Graph 2: Histogram for minutes of Sleep / no of days 
plt.xlabel('minutes of sleep')
plt.ylabel('no of days')
plot2=plt.hist(workout_data['minutes(slept)'],bins=5,label='Sleeping Time',color='red',rwidth=0.95)
```

```python
# Conclusion: A person sleeps 250 to 350 minutes per day  in maximum noi  of the day.
```

```python
#Graph 3: sleep Quality over time
plt.xlabel('minutes(slept)')
plt.ylabel('Activity (steps)')
plot3=sb.scatterplot(workout_data['minutes(slept)'],workout_data['Activity (steps)'], color='green')
```

```python
# Conclusion: There is more of activity during 250 to 450 minutes of sleep. 
```

```python
#Graph 4 :Duration mean Quality
plt.xlabel('minutes(slept)')
plt.ylabel('Sleep quality')
plot4= sb.scatterplot(workout_data['minutes(slept)'],workout_data['Sleep quality'],color='orange')
```

```python
# Conclusion: According to the graph the quality of sleep  does not change over duration of sleep.even during the initial stage of sleep quality is high.  
```
