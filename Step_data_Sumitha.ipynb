```python
import numpy as np
import pandas as pd
from datetime import datetime
import time
import matplotlib.pyplot as plt
import statistics as s
import seaborn as sb
```

```python
#Graph 1:Calories burnt with Distance 
cols = ['#EE7550', '#F19463', '#F6B176']
plt.scatter(step_data.distance, step_data.calorie, facecolors='white', edgecolors='0', s = 5, lw = 0.7, marker='o')
plt.axvline(step_data.distance.mean(), color='k', linestyle='dashed', linewidth=1)
plt.axhline(step_data.calorie.mean(), color='k', linestyle='dashed', linewidth=1)
plt.text(42,2,'distance',rotation=90)
plt.text(0,3,'calorie',rotation=0)
plt.xlabel('distance')
plt.ylabel('calorie')
plt.title('Calories burnt with Distance ')
```

```python
# Conclusion : From the graph it is evident that average a person walks around 40metres to burn approximately 3 calories.
#/ It is also evident that walking more distance also burns more calories.
```

```python
#Graph2: calories burnt and step count
total_count=step_data['calorie'].groupby([step_data['count']]).mean()
plt.figure(figsize=(40,20))
plt.title('calories burnt')
total_count.plot(kind='bar')

```

```python
#Conclusion:From the Graph it is evident that more the stepcount more calories are burnt.so person can burn more calories by
#/ taking more step or walking more.
```
